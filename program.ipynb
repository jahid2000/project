```python
def longest_subtring(X, Y):
    X = X.lower()
    Y = Y.lower()
    X = X.split(" ")
    Y = Y.split(" ")
    m,n = len(X), len(Y)
    LCSuff = [[0 for k in range(n+1)] for l in range(m+1)] 
    result = 0 
    for i in range(m + 1): 
        for j in range(n + 1): 
            if (i == 0 or j == 0): 
                LCSuff[i][j] = 0
            elif (X[i-1].strip() == Y[j-1].strip()): 
                LCSuff[i][j] = LCSuff[i-1][j-1] + 1
                result = max(result, LCSuff[i][j]) 
            else: 
                LCSuff[i][j] = 0
    return result
longest_subtring("Ab cd", "ax cd")           # prints 2
```
